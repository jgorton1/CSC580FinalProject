```python
import os
import pandas as pd
```

```python
csv_file_path = 'features_3_sec.csv'
df = pd.read_csv(csv_file_path)
```

```python
train_directory = "train"#"C:/Users/Julius Gorton/csc580/CSC580FinalProject/Data/train"
train_file_names = []

for root, dirs, files in os.walk(train_directory):
    for file in files:
        file_path = os.path.join(root, file)
        train_file_names.append(os.path.basename(file_path))
```

```python
#train_file_names = [filename[:-4] for filename in train_file_names]
train_file_names
```

```python
import re
train_df = df[df['filename'].str.contains('|'.join(map(re.escape, train_file_names)))]
```

```python
train_df
```

```python
train_df.to_csv('train_data_3sec.csv', index=False)
```
